# Hello" 500 times in Python, along with a hint for the question:

Hint for the Question:
To print a string multiple times without manually copying and pasting, you can use loops or Python's string multiplication feature. A loop is more flexible and scalable for such tasks.

```python
count = 1

while count <= 500:
    print("Hello")
    count += 1
```

```python
i = 1
while i <= 500:
    print("Hello", i)
    i += 1
```

# Search for a Number x in a Tuple Using a Loop
[1, 4, 9, 16, 25, 36, 49, 64, 81, 100]


```python
nums = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)
x = 64
idx = 0

while idx < len(nums):
    if nums[idx] == x:
        print("Number found at index", idx)
    else:
        print("Number not found")
    idx += 1
```

```python
nums = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)
x = 64
idx = 0

while idx < len(nums):
    if nums[idx] == x:
        print("Number found at index", idx)
        break
    idx += 1
```

# Continue the loop after finding the first occurrence of a number

```python
i = 0
while i < 10:
    if i == 5:
        i += 1
        continue
    print(i)
    i += 1
```

```python
i = 0
while i < 10:
    if i % 2 == 0:
        i += 1
        continue
    print(i)
    i += 1
```

```python
i = 0
while i < 10:
    if (i % 2 != 0):
        i += 1
        continue
    print(i)
    i += 1
```
